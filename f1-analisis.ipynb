```python
#Librerias
import requests
import pandas as pd
import json
import warnings
warnings.filterwarnings("ignore")
```

```python
#Obtenemos nuestra base de datos
url = "https://ergast.com/api/f1/2022/22/results.json"
```

```python
#Se genera una variable utilizando la paqueteria request, obteninedo la info de la pagina de internet, se extraen en forma 
#datos json
response = requests.get(url)
content = json.loads(response.content)
```

```python
#Creamos un data frame con los resultados de la carrera 22 del año 2022 
results = pd.DataFrame(content["MRData"]['RaceTable']['Races'][0]['Results'])
```

```python
results.head()
```

```python
#Extraemos los nombres de los conductores
for i in range(len(results["Driver"])):
    print(results["Driver"][i]["driverId"])
```

```python
#Agregamos directamente el nombre der conductor y el constructor en el data frame
results["Driver"] = results["Driver"].apply(lambda x: x["driverId"])
results['Constructor'] = results['Constructor'].apply(lambda x: x['constructorId'])
results['FastestLap'] = results['FastestLap'].apply(lambda x: x['Time'])
```

#### TABLA RACES

```python
#Diccionario vacio para tabla races
races= {'season':[],
       'round' : [],
       'raceName':[],
       'circuitId':[],
       'lat':[],
       'long':[],
       'country':[],
       'date':[],
       'url':[]}
```

```python
#Contenido de races
content["MRData"]['RaceTable']['Races'][0]
```

```python
#Carreras de  temporadas 2010-2022
for year in list (range(2010,2023)):
    url = "https://ergast.com/api/f1/{}.json"
    response = requests.get(url.format(year))
    content = json.loads(response.content)
    
    for item in content["MRData"]['RaceTable']['Races']:
        races['season'].append(int(item['season']))
        races['round'].append(int(item['round']))
        races['raceName'].append(item['raceName'])
        races['circuitId'].append(item['Circuit']['circuitId'])
        races['lat'].append(item['Circuit']['Location']['lat'])
        races['long'].append(item['Circuit']['Location']['long'])
        races['country'].append(item['Circuit']['Location']['country'])
        races['date'].append(item['date'])
        races['url'].append(item['url'])
    
```

```python
#Tabla races
pd.DataFrame(races).to_csv("races.csv", index = False)
```

```python
url = "https://ergast.com/api/f1/seasons.json"
response = requests.get(url.format(2022))
content = json.loads(response.content)
```

```python
#Diccionario vacio
seasons= {'season':[],
       'url':[]}
```

```python
#Contenido de season
content["MRData"]
```

```python
for item in content["MRData"]['SeasonTable']['Seasons']:
        seasons['season'].append(int(item['season']))
        seasons['url'].append(item['url'])
```

```python
pd.DataFrame(seasons)
```

```python
df_results_2022=pd.DataFrame()
for race in list(range(1,23)):
    url = "https://ergast.com/api/f1/2022/{}/results.json"
    response = requests.get(url.format(race))
    content = json.loads(response.content)
    df_results_2022=df_results_2022.append(content["MRData"]['RaceTable']['Races'][0]['Results'])
```

```python
df_results_2022['Driver']=df_results_2022['Driver'].apply(lambda x: x['driverId'])
```

```python
#Posiciones
df_results_2022['points']=df_results_2022.points.astype(int)
df_results_2022
```

```python
#Posiciones 
df_results_2022.groupby(['Driver']).agg({'points':'sum'}).sort_values(by ='points', ascending=False)
```
